7) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.(155 dahil, 180 dahil değil)

not1 : Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.

not2 : horsepower sütunu içinde '?' şeklinde değerler var. Bu değerleri önce nan haline getirip devamında horsepower sütununu float hale getirip istenen sorguyu yapabilirsiniz.

not3 : nan değerler varken bile sütunun dtype'ını 'float' yapabilme sebebimiz nan değerlerin arka planda içerisinde rastgele anlamsız float bir sayı tutmasıdır.

```python
df.info()
```

```python
df['horsepower'].replace('?' , np.nan , inplace = True)

df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.info()
```

```python
df[df['horsepower'].between(155,180 , inclusive = 'left')]
```

8) Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından df'in indexlerini sıfırlayın.

```python
df.sort_values('horsepower' , ascending = False).reset_index(drop = True)
```

9) horsepower sütunununu float dtype'ına döndürmek için '?' değerlerini nan'a döndürmüştük ancak öyle bırakmıştık. Şimdi bu nan değerleri doldurma vakti.

Basitçe horsepower sütununun ortalamasıyla horsepower içindeki nan değerleri doldurmaya çalışın. 

Devamında tüm sütunları nan değer kaldı mı diye kontrol edin ve ondan sonra df'i yazdırın.

```python
df['horsepower'].fillna(df['horsepower'].mean() , inplace = True)
```
