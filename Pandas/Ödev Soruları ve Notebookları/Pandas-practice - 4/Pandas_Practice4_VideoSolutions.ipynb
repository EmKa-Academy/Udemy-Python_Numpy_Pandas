```python
fill_cols = list(df.isna().sum()[df.isna().sum() != 0].index)
```

```python
# 'Region' sütununun ortalamalarını hesapla
region_means = df.groupby('Region').mean()

# NaN değerleri 'Region' ortalamalarıyla doldur
for column in fill_cols:
    df[column] = df.groupby('Region')[column].transform(lambda x: x.fillna(x.mean()))
```
