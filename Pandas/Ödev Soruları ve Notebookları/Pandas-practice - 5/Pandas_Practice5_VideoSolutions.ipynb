9) production_df içinde de 'Tarih' sütunu bulunuyor ve o tarihlerdeki kullanılan benzin miktarı , gaz miktarı ve daha pek çok bilgidne oluşan bir dataframe olduğunu gördük.

Buradaki bilgileri bizim orjinal df'imiz olan train_df'e sütun bilgisi olarak ekleyebiliriz. Bu sayede problemimiz TimeSeries probleminden bir nebze çıkıp Regression problemine döner ve işimiz biraz daha kolaylaşır.

train_df içindeki Tarihler için gerekli bilgileri production_df'den çekerek df'leri birleştirmeye çalışın. Özetle train_df'iniz ile production_df'i merge edip featured_df adında yeni bir df'e eşitleyip featured_df'i ekrana yazdırın.

```python
featured_df = pd.merge(left = train_df , right = production_df , on='Tarih')
```

```python
print(f'Featured DF Shape {featured_df.shape}')
print(f'Train DF Shape {train_df.shape}')
print(f'Production DF Shape {production_df.shape}')
```
