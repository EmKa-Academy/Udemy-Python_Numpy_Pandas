# Concatenate(Birleştirme)

```python
import numpy as np
import pandas as pd
```

```python
# Concatenate işlemi basitçe birleştirmedir.
# Dataframeleri alt alta ya da yan yana birleştirmekten ibarettir.
```

```python
data_1 = {'isim' : ['Emincan' , 'Aslı' , 'Derya' , 'Burak'] , 'yas' : [26,24,39,21]}
data_2 = {'isim' : ['Onur' , 'Kerem' , 'Mert' , 'Pelin'] , 'yas' : [20,16,34,30]}
data_3 = {'il' : ['Rize' , 'İstanbul' , 'Ordu' , 'Ankara'] , 'meslek' : ['Öğretmen','Doktor','Bakıcı','Yazılımcı']}
```

```python
df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)
df_3 = pd.DataFrame(data_3)
```

```python
df_1
```

```python
df_2
```

```python
pd.concat([df_1 , df_2] , axis = 0).reset_index(drop=True)
```

```python
pd.concat([df_1 , df_2] , axis = 1) # Mantıksız
```

```python
pd.concat([df_1 , df_3] , axis = 1)
```
