# Eksik Veriler(Missing Valueler)

```python
# Missing Value, 'Eksik Veri' kavramı datamız içinde bilinmeyen verileri ifade eder.
# Missing valuelar farklı formatlarda görünebilir(nan , NA , NaT , vb.)
```

```python
# Missing Valuelarla ilgili yapılabilecek 3 işlem vardır.
# 1) Hiçbir şey yapmayabiliriz. (Analizlerde yapılabilir)
# 2) Mantıklı değerler ile doldurulabilir. (.fillna())
# 3) Eksik verileri satır ya da sütun bazlı düşürebiliriz. (.dropna())
```

```python
df.dropna() # Bir satırda herhangi bir nan değer varsa o satırı düşür. (Bu çok kullanılmaz.)
```

```python
df.dropna(thresh = 1) # thresh eşik(sınır) demek. Satırımızda 1 tane nan olmayan değer varsa o satırı tut.
```

```python
df.dropna(axis = 0, thresh = 5) # Satırımızda 5 tane nan olmayan değer varsa o satırı tut.
```

```python
df.head()
```

```python
df.dropna(axis = 1)
```

```python
df.dropna(axis = 1 , thresh = 3)
```

```python
df.dropna(axis = 1 , thresh = 4)
```

```python
df.head()
```

```python
df.dropna(subset = ['pre_movie_score']) # pre_movie_score'u nan olan satırları düşür
```

```python
df.dropna(subset = ['first_name'])
```

```python
df.dropna(subset = ['first_name' , 'pre_movie_score']) # first_name veya pre_movie_score sütunlarından herhangi biri nan ise o satırı düşür.
```

```python
# Fillna(Doldurma)
```

```python
df.head()
```

```python
df.dropna(how = 'all' , inplace = True)
```

```python
df.head()
```

```python
df['pre_movie_score'].mean()
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean() , inplace = True)
```

```python
df.head()
```

```python
df['post_movie_score'].fillna(5 , inplace = True)
```

```python
df.head()
```

```python
df.reset_index(inplace = True , drop=True)
```
