```python
# .groupby(sütun_ismi) methodu ile bir sütundaki değerlere göre gruplama yapılabilir. Ardından bir aggregate function kullanılarak gruplar incelenebilir.
```

Common Options:

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values
    
Full List at the Online Documentation: https://pandas.pydata.org/docs/reference/groupby.html

```python
df['horsepower'].replace('?' , np.nan , inplace = True)
```

```python
df[df['horsepower'].isna()]
```

```python
df['horsepower'] = df['horsepower'].astype('float64')
```

```python
df.info()
```

```python
df.groupby(['model_year' , 'cylinders'])['horsepower'].mean()
```

```python
df.groupby(['model_year' , 'cylinders'])['horsepower'].mean().loc[(71,4)]
```

```python
df.loc[(df['model_year'] == 71) & (df['cylinders'] == 4) , 'horsepower'] = df.loc[(df['model_year'] == 71) & (df['cylinders'] == 4) , 'horsepower'].fillna(77.5)
df.loc[(df['model_year'] == 74) & (df['cylinders'] == 6) , 'horsepower'] = df.loc[(df['model_year'] == 74) & (df['cylinders'] == 6) , 'horsepower'].fillna(101.5)
df.loc[(df['model_year'] == 80) & (df['cylinders'] == 4) , 'horsepower'] = df.loc[(df['model_year'] == 80) & (df['cylinders'] == 4) , 'horsepower'].fillna(74)
df.loc[(df['model_year'] == 81) & (df['cylinders'] == 4) , 'horsepower'] = df.loc[(df['model_year'] == 81) & (df['cylinders'] == 4) , 'horsepower'].fillna(73)
df.loc[(df['model_year'] == 82) & (df['cylinders'] == 4) , 'horsepower'] = df.loc[(df['model_year'] == 82) & (df['cylinders'] == 4) , 'horsepower'].fillna(79)
```
