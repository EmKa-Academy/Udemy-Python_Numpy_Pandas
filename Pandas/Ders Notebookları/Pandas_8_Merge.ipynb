# Merge

```python
import numpy as np
import pandas as pd
```

```python
# Merge işlemi 4 adettir. (inner , outer , left , right)
# SQL'deki join işlemine benzer bir mantıkta df'leri birleştirmeye yarar.
# Sütun bazlı birleştirme yapılır. Ortak bir sütun olması gerekir.
```

```python
data_1 = {'kayıt_id' : [1,2,3,4] , 'isim' : ['Onur','Burak' , 'Pelin' , 'Ilgın']}
data_2 = {'giris_id' : [1,2,3,4] , 'isim' : ['Alper','Ilgın' , 'Onur' , 'Emincan']}
```

```python
df1 = pd.DataFrame(data_1)
df2 = pd.DataFrame(data_2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(how = 'inner' , left = df1 , right = df2 )
```

```python
pd.merge(how = 'left' , left = df1 , right = df2 )
```

```python
# pd.merge(how = 'left' , left = df2 , right = df1 )

pd.merge(how = 'right' , left = df1 , right = df2 )
```

```python
pd.merge(how = 'outer' , left = df1 , right = df2 )
```

```python
df1
```

```python
df2.rename(columns={'isim' : 'name'} , inplace = True)
```

```python
df1
```

```python
df2
```

```python
pd.merge(how = 'inner' , left = df1 , right = df2 , left_on= 'isim' , right_on = 'name')
```
