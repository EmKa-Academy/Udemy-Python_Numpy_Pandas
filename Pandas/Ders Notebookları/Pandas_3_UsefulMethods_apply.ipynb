# .apply() Methodu

```python
# 'apply' kelime anlamı olarak 'uygulamak' anlamına gelir.
# .apply() methodunun arka planında for döngüsü çalışır.
# .apply() methodunu bir serideki ya da bir dataframe'in bir sütunundaki bütün elemanlara aynı fonksiyonu uygulamak için kullanırız.
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

Amaç : tip sütunundaki bütün değerlerini 2 katını almak

# 1. Yol

```python
def kareAl(x):
  return x**2
```

```python
# df['tip'] = df['tip'].apply(kareAl)
df['tip'].apply(kareAl)
```

# 2. Yol

```python
df['tip'].apply(lambda x : x**2)
```

# 3.Yol Vectorized(En Hızlısı)

```python
df['tip'] ** 2
```

Amaç : 'CC Number' sütunundaki her değerin son 4 rakamını çekme

```python
df.head()
```

```python
df.info()
```

```python
df['CC Number'] = df['CC Number'].astype(str)
```

```python
df.head()
```

Amaç : 'CC Number' sütunundaki her değerin son 4 rakamını çekme

# 1. Yol

```python
def son4rakam(x):
  return x[-4:]
```

```python
# df['CC Number'] = df['CC Number'].apply(son4rakam)
df['CC Number'].apply(son4rakam)
```

# 2.Yol

```python
df['CC Number'].apply(lambda x : x[-4:])
```

# 3.Yol Vectorized

```python
df['CC Number'].str[-4:]
```

# Apply Methodunu birden fazla sütuna uygulama

# 1.Yol

```python
df.head()
```

```python
def kareAl(x):
  return x**2
```

```python
df[['total_bill' , 'tip']].apply(kareAl)
```

# 2.Yol

```python
df[['total_bill' , 'tip']].apply(lambda x : x**2)
```

# 3.Yol

```python
# df[['total_bill' , 'tip']] = np.square(df[['total_bill' , 'tip']])
np.square(df[['total_bill' , 'tip']])
```
