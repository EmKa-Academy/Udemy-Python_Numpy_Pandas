# Aşağıdaki UsefulMethodları Göreceğiz:
* describe() : Bütün sütunların istatistiksel özetini çıkartır
* value_counts() : Bir sütundaki unique değerlerin her birinden kaç tane olduğunu gösterir
* unique() : Bir sütunda hangi eşsiz değerler bulunuyor
* nunique() : Bir sütunda toplam kaç tane eşsiz değer var
* nlargest() : Bir sütundaki en büyük n tane değeri getirmeye yarar.
* nsmallest() : Bir sütundaki en küçük n tane değeri getirmeye yarar.
* sort_values() : Bir sütundaki değerlere göre satırları sıralamaya yarar.
* corr() : Sütunlar arasındaki korelasyona(ilişkiyi) hesaplar.
* map() : .apply() methodunun aynısı, sadece kullanım alanları farklı.
* replace() : 2 parametre alır. 1.'si değiştirilecek ifade, 2.si ne ile değiştireleceği.
* duplicated() : Her şeyleri aynı duplicated satırları getirir.
* between() : Bir sütun içindeki değerlerden istenen aralıktakileri çekmek için kullanılır.
* argmax() : Bir sütundaki maksimum değerin kaçıncı indexde bulunduğunu döndürür.(idxmax)
* argmin() : Bir sütundaki minumum değerin kaçıncı indexde bulunduğunu döndürüyor.(idxmin)
