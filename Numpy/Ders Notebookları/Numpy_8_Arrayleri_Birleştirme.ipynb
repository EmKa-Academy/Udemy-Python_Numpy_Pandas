### 2D arrayler

```python
arr1 = np.random.randint(low=0, high=5, size=(3,3))
arr1
```

```python
arr2 = np.random.randint(low=0, high=5, size=(3,3))
arr2
```

```python
np.concatenate([arr1, arr2], axis=0) # (6, 3)
```

```python
np.concatenate([arr1, arr2], axis=1) # (3, 6)
```

```python
np.stack([arr1, arr2], axis=0) # (2, 3, 3)
```

```python
np.stack([arr1, arr2], axis=1) # (3, 2, 3)
```

```python
np.stack([arr1, arr2], axis=0)[0]
```
