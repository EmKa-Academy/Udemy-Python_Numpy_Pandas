```python
A = np.array([
    [4, -1, 1],
    [-1, 1, 1],
    [1, 2, 2]
])
# bilinmeyenler = [x, y, z]
b = np.array([2, 2, 3])
```

```python
np.linalg.det(A)
```

```python
x = np.linalg.inv(A) @ b
x
```

```python
np.linalg.solve(A, b)
```
