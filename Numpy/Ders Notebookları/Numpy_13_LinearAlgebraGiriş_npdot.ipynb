```python
import numpy as np

np.__version__
```

```python
# 1D arrayler.
arr1 = np.array([1, 2, 3, 4]) # shape (4,)
arr2 = np.array([5, 6, 7, 8])
```

```python
np.sum(np.multiply(arr1, arr2))
```

```python
np.dot(arr1, arr2)
```

```python
np.dot(arr2, arr1)
```

```python
arr3 = np.array([1, 2]) # shape = (2,)
np.dot(arr2, arr3)
```

```python
# 2D arrayler.
arr4 = np.array([[1, 2], 
                 [3, 4]]) # shape (2, 2)
arr5 = np.array([[5, 6],
                 [7, 8]])
```

```python
np.dot(arr4, arr5)
```
