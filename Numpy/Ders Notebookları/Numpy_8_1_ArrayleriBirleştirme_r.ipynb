```python
import numpy as np
import perfplot

np.__version__
```

```python
np.concatenate([
    np.array([0.0]), np.array([1.0])
])
```

```python
np.concatenate([
    0.0, np.array([1.0])
])
```

```python
np.r_[0.0, np.array([1.0])]
```

```python
np.concatenate([
    np.arange(0, 10), np.array([1.0])
])
```

```python
np.r_[slice(1, 5), np.array([1.0])]
```

```python
benchmark = perfplot.bench(
    kernels = [
        lambda b: np.r_[b, b],
        lambda b: np.concatenate([b, b])
    ],
    n_range = [2**b for b in range(20)],
    setup = np.random.rand,
    labels = ["np.r_", "np.concatenate"],
    xlabel = "Array boyutu"
)
```

```python
benchmark.show()
```
