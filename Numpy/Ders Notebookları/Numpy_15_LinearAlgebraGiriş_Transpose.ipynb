```python
import numpy as np

np.__version__
```

```python
arr1 = np.array([[1, 2, 3], [4, 5, 6]]) #shape = (2, 3)
arr1
```

```python
# Tranpose işlemi 2D arraylerde, satır ve sütunların yer
# değiştirmesidir. Neticede shape --> (3, 2) olacaktır (bu durumda)
```

```python
# 1 4
# 2 5
# 3 6
```

```python
arr1.transpose()
```

```python
arr1.shape
```

```python
np.transpose(arr1, axes = [1, 0])
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]]) # (2, 3)
arr2
```

```python
# (2, 3) - (2, 3)
# İkinciyi tranpose edersek (2, 3) - (3, 2)
# matmul yapılabilir. Çıktı shape'i (2, 2) matris
```

```python
arr1 @ arr2.transpose()
```

```python
# (224, 224, 3) - 3 channel size RGB olduğunu belirtir.
# np.transpose(array, axes = [2, 0, 1]) --> çıktı shape'i (3, 224, 224)
```
