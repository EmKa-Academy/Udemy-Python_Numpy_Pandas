```python
arr = np.array([
    [0, 1, 2],
    [3, 4, 5],
    [6, 7, 2],
    [9, 10, 11],
    [12, 13, 14]
])
arr
```

```python
arr.shape
```

```python
# İlk iki satır ve ilk sütundaki değerler
arr[0:2, 0]
```

```python
# Tüm satırları ve ikinci sütundan sonraki değerler
arr[:, 2:]
```

```python
# İlk iki sütunu ve tüm satırları içeren indeksleme
arr[:, :2]
```

```python
# İlk satırdan ikinci satıra kadar olan (ikinci satır dahil) satırlar ve ilk sütundan üçüncü sütuna kadar
arr[:2, :]
```

```python
arr[:2, [0, 2]]
```
