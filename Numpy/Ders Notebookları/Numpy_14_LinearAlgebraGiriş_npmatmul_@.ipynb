```python
import numpy as np

np.__version__
```

```python
arr1 = np.random.uniform(low = -1, high = 1, size = (20, 2))
arr2 = np.random.uniform(low = -1, high = 1, size = (2, 3))
```

```python
# Matris çarpımı yaparken
# arr1 shape = (a, b) - a = 20, b = 2
# arr2 shape'i = (b, c) - b = 2, c = 3
# çıktı shape'i = (a, c) --> a matmul b
```

```python
matmul_result = np.matmul(arr1, arr2)
```

```python
operator_result = arr1 @ arr2
```

```python
np.allclose(matmul_result, operator_result)
```

```python
arr1.__matmul__(arr2)
```

```python
np.dot(arr1, arr2)
```
