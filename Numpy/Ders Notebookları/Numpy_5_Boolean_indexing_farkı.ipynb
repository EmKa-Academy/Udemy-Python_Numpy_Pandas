```python
import numpy as np
import time

np.__version__
```

```python
array_shape = (100, 100)

arr = np.random.randint(0, 1000, size = array_shape)
arr_copy = arr.copy()
```

```python
arr.shape
```

### 3'e bölünenilen sayıları -1 ile değiştirmek

```python
%%timeit
for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i, j] % 3 == 0:
            arr[i, j] = -1
```

```python
%%timeit
arr_copy[arr_copy % 3 == 0] = -1
```

```python
(arr_copy == arr).all()
```
