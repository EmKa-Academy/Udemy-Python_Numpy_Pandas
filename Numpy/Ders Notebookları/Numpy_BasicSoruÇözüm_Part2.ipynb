16) (4, 1) shape olan iki array oluşturup sütun bazında concat ederek (4, 1, 2) shape'inde array döndürün.

````python
arr1 = np.expand_dims(np.array([1, 2, 3, 4]), axis=-1)
arr2 = np.array([5, 6, 7, 8])[:, np.newaxis]
arr1
````

````python
arr2
````

````python
arr2.shape
````

````python
arr = np.concatenate([arr1, arr2], axis=1)
arr
````

````python
arr = np.expand_dims(arr, axis=1)
arr
````

````python
arr.shape
````

````python
arr = np.dstack((arr1, arr2))
````

````python
arr[0]
````

17) 1'den 9'a kadar olan sayılardan (9 tane item içeren) array oluşturup 3 eşit parçaya bölün.

````python
numbers = np.arange(0, 8)
````

````python
for _ in range(5):
    print(np.random.choice(numbers))
````

````python
arr = np.array([np.random.choice(numbers) for _ in range(9)])
arr
````

````python
np.split(arr, [3, 6])
````
