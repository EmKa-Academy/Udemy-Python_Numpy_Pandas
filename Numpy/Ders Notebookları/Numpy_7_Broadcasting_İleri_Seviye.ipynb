```python
import numpy as np
np.__version__
```

```python
exam_scores = np.random.uniform(low = 0, high = 101, size = (20, 4))
exam_scores = exam_scores.astype(np.int32)
exam_scores
```

```python
exam_scores_normalized = exam_scores / 100
exam_scores_normalized
```

```python
exam_scores_normalized.shape
```

```python
mean_exam_scores = np.mean(exam_scores_normalized, axis=0)
mean_exam_scores = np.around(mean_exam_scores, decimals=2)
mean_exam_scores
```
