6) 7x7 bir array oluşturup içinde bulunan 5x5'lik kısımların değerlerini 2 olacak şekilde değiştirin.

```python
arr = (np.random.uniform(low=1, high=5, size = (7, 7)) * 50).astype(np.int32)
arr 
```

```python
arr[1:-1, 1:-1] = 2
arr
```
