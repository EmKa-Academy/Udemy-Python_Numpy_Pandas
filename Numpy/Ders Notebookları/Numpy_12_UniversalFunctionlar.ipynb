```python
import numpy as np

np.__version__
```

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
```

```python
l1 + l2
```

```python
l_toplam = []
for l1_item, l2_item in zip(l1, l2):
    l_toplam.append(l1_item + l2_item)
```

```python
l_toplam
```

```python
l_toplam = np.add(l1, l2)
l_toplam
```

```python
np.add.reduce(l_toplam)
```

```python
np.sum(l_toplam)
```
